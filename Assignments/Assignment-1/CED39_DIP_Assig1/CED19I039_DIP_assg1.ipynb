## Part1 - RMSE plotting using Color Images  

```python
#importing necessary libraries
import glob
import math
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=10,10
import matplotlib.image as mpimg
```

```python
#Input list of images from path
images = [cv2.imread(file) for file in glob.glob("dip1_dataset/y*.jpg")]
images=np.array(images)
## 3 channel images, colored
print(images.shape)
```

```python
#List containg image numbers
img_lst = list(range(1,len(images)))
img_lst=np.array(img_lst)
img_lst
```

```python
#Verifying no. of input images
l=len(images)
l
```

```python
#Ploting all input images
for i in range(l):
    plt.subplot(10,10,i+1),plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    #plt.title(i)
    plt.xticks([]),plt.yticks([])
plt.show()
print(l,"- Images")
```

```python
#Considering 1st input image as reference
ref_image=images[0];
```

```python
#Calculating rmse score for each image w.r.t reference image
rmse_score=[]
for i in range(1,l):
    rmse=math.sqrt(np.square(np.subtract(ref_image,images[i])).mean())
    rmse_score.append(rmse)
rmse_score=np.array(rmse_score)
print(rmse_score)
```

```python
#Rounding off values to two decimal places
rmse_score=np.round_(rmse_score, decimals=2)
print(rmse_score)
```

```python
# Plot the values in the graph with the indexed images in the X-axis and RMSE score in the Y-axis.
plt.title("Image number vs RMSE score", fontsize=15)
plt.xlabel("Image number",fontsize=15)
plt.ylabel("RMSE score",fontsize=15)
plt.plot(rmse_score)
"""
plt.plot(img_lst,rmse_score)

##forcing integer x-axis
x=img_lst
new_list = range(math.floor(min(x)), math.ceil(max(x))+1)
plt.xticks(new_list)
"""
plt.show()
```

## Part2 - RMSE plotting using Grayscale Images  

```python
#Converting all images to grayscale
gray_images=[]
for i in range(l):
    gray=cv2.cvtColor(images[i], cv2.COLOR_BGR2GRAY)
    gray_images.append(gray)
gray_images=np.array(gray_images)
##Single channnel images, grayscale
print(gray_images.shape)
```

```python
#Ploting all input images
for i in range(len(images)):
    plt.subplot(10,10,i+1),plt.imshow(gray_images[i],cmap='gray', vmin=0, vmax=255)
    #plt.title(i)
    plt.xticks([]),plt.yticks([])
plt.show()
print(l,"- Images")
```

```python
#Considering 1st input image as reference
gray_ref_image=gray_images[0];
```

```python
#Calculating rmse score for each image w.r.t reference image
gray_rmse_score=[]
for i in range(1,l):
    gray_rmse=math.sqrt(np.square(np.subtract(gray_ref_image,gray_images[i])).mean())
    gray_rmse_score.append(gray_rmse)
gray_rmse_score=np.array(gray_rmse_score)
print(gray_rmse_score)
```

```python
#Rounding off values to two decimal places
gray_rmse_score=np.round_(gray_rmse_score, decimals=2)
print(gray_rmse_score)
```

```python
# Plot the values in the graph with the indexed images in the X-axis and RMSE score in the Y-axis.
plt.title("Image number vs RMSE score(grayscale)", fontsize=15)
plt.xlabel("Image number",fontsize=15)
plt.ylabel("RMSE score(grayscale)",fontsize=15)
#plt.plot(img_lst,gray_rmse_score)
plt.plot(gray_rmse_score)
##forcing integer x-axis
"""
x=img_lst
new_list = range(math.floor(min(x)), math.ceil(max(x))+1)
plt.xticks(new_list)
"""
plt.show()
```

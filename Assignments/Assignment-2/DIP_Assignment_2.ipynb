## Libraries


```python
from matplotlib import pyplot as plt                                              
import cv2
import glob
import os
import numpy as np
import ntpath
import math
```

## Read Lena Image 

```python
cwd = os.getcwd()                                            # get current directory
img = cv2.imread(cwd+"/Lena.png") 
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)             # read image for its location
```

# Part 1

## Down sample the image

```python
gray_img_128=cv2.resize(gray_img,(128,128)) 
gray_img_64=cv2.resize(gray_img,(64,64)) 
gray_img_32=cv2.resize(gray_img,(32,32)) 
gray_img_16=cv2.resize(gray_img,(16,16)) 
```

## Display images

```python
f = plt.figure()
f.set_figwidth(6*5)
f.set_figheight(1*5)

#for i in range(10):
    # display Different Angle
ax = plt.subplot(1, 6, 1)
plt.imshow(img[:, :, ::-1])
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

ax = plt.subplot(1, 6, 2)
plt.imshow(gray_img, cmap='gray', vmin = 0, vmax = 255)
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

ax = plt.subplot(1, 6, 3)
plt.imshow(gray_img_128, cmap='gray', vmin = 0, vmax = 255)
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

ax = plt.subplot(1, 6, 4)
plt.imshow(gray_img_64, cmap='gray', vmin = 0, vmax = 255)
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

ax = plt.subplot(1, 6, 5)
plt.imshow(gray_img_32, cmap='gray', vmin = 0, vmax = 255)
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)


ax = plt.subplot(1, 6, 6)
plt.imshow(gray_img_16, cmap='gray', vmin = 0, vmax = 255)
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)



plt.figtext(0.16,0.9, "Original", ha="center", va="top", fontsize=14, color="r")
plt.figtext(0.33,0.9, "Grayscale", ha="center", va="top", fontsize=14, color="r")
plt.figtext(0.45,0.9, "128*128", ha="center", va="top", fontsize=14, color="r")
plt.figtext(0.60,0.9, "64*64", ha="center", va="top", fontsize=14, color="r")
plt.figtext(0.72,0.9, "32*32", ha="center", va="top", fontsize=14, color="r")
plt.figtext(0.85,0.9, "16*16", ha="center", va="top", fontsize=14, color="r")

plt.subplots_adjust(hspace = 0.5 )
plt.show()
```
